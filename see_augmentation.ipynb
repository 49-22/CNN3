```python
import os
import glob
from shutil import copyfile
if not os.path.exists("augment/dogs"):
    os.makedirs("augment/dogs")
d = glob.glob("../dogcat-data/train/dogs/*")[0]
print(d)
copyfile(d, "augment/dogs/"+ d.split("/")[-1])
```

```python
train_iter = train_datagen.flow_from_directory(
    "augment",
    target_size=(100, 100),
    batch_size=9,
    class_mode="categorical"
)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (10,10)
for i in range(9):
    images, labels = next(train_iter)
    plt.subplot(3,3,i+1)
    plt.imshow((images[0] - np.min(images[0])) / np.ptp(images[0]), interpolation='none')
    plt.title("label {}".format(labels[0]))
```
